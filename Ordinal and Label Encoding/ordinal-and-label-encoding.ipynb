```python
data = pd.read_csv("/kaggle/input/customer/customer.csv")
data
```

```python
#For now we will only include the categorical data so we will remove numerical data
#The gender column will be required to apply one hot encoding so for now we will also remove that
data = data.iloc[:,2:]
data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.iloc[:,:2],data.iloc[:,2],test_size=0.2)
```

**Ordinal Encoding**<br>
Do this if you know that the data has some kind of order

```python
#Now ww wil do Ordinal encoding
#The review and education column is ordinal data
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder(categories=[["Poor","Good","Average"],["School","UG","PG"]])
ord_enc.fit(X_train)
```

```python
#Remember train with train_data and apply to bot train and test
X_train = ord_enc.transform(X_train)
X_test = ord_enc.transform(X_train)

#Just ignore the error
```

```python
#The output will be numpy arrray you can convert to dataframe and merge that data or you can use column transformer
X_train = pd.DataFrame(X_train,columns=["review","education"])
```
