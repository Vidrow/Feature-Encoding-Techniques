**OHE using sklearn**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.iloc[:,:4],data.iloc[:,4],test_size=0.2)

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(drop="first",sparse=False,dtype=int)
ohe.fit(X_train[["fuel","owner"]])

#sparse=False will give as array rather than sparse matrix
```

```python
new_X_train = ohe.transform(X_train[["fuel","owner"]])
```

```python
new_X_test = ohe.transform(X_test[["fuel","owner"]])
```
